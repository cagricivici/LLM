```python
# Importing Libraries
import os
import requests # sending http requests
from dotenv import load_dotenv # fetching the .env file
from bs4 import BeautifulSoup # clean up the web page content
from IPython.display import display, Markdown
from openai import OpenAI
```

```python
# Connecting to OpenAI
dotenv_path = "/content/drive/MyDrive/calismalar/PythonStudy/summary file/.env"
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
  print('API key not found')
else:
  print('Succesfull')

client = OpenAI(api_key=api_key)
```

```python
openai  = OpenAI()
```

```python
class Website:
  """
  this class will reach out url, title and text in body from a html page.
  """
  def __init__(self, url: str):
    self.url = url

    #fetching and parsing and html page
    response = requests.get(url) # sending the http request with url
    soup = BeautifulSoup(response.content, 'html.parser')

    self.title = soup.title.string if soup.title else 'No Title Found'
    self.text = soup.body.get_text(separator='\n',strip = True) if soup.body else 'No Text Found'
```

```python
def user_prompt_for(website):
  '''
  This class will send the user prompt to openai by gathering all informations
  '''
  user_prompt = f"""
website is {website.url} url.
the title is {website.title}
The content is as below:
please provide a short summary of this website in markdown. .
If it includes news or announcements, then summarize those too.
website content: {website.text}
"""
  return user_prompt
```

```python
deneme = 'https://read.cv/cagri'
website = Website(deneme)
prompt = user_prompt_for(website)
print(prompt)
```

It seems that user prompts can pack all regarding info. that's fine right now. then i will put it into message dict in openai.

```python
system_prompts = 'anaylize the content of the website then please provide a short summary'

def message_for(website):
  """
  this function will pack message packs so taht user prompts and system prompts can be sent to openai
  """
  user_prompt = user_prompt_for(website) #gathered info will be called here then use it.
  message = [
      {'role': 'system', 'content': system_prompts},
      {'role': 'user', 'content': user_prompt}
  ]
  return message
```

```python
def summarize(url):
  website = Website(url) #URL, Text and Title is ready to use
  message = message_for(website) #info will be gathered here then put into message dict

  response = openai.chat.completions.create(
        model="gpt-4",
        messages= message)
  return response.choices[0].message.content
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
url = 'https://read.cv/cagri'
display_summary(url)
```
